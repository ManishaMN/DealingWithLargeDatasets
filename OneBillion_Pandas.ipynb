```python
#Tip5: Loading data using chunksize option and performing some calculations on this chunk using a function

# obtain cities with max temperature in that chunk, sort cities in ascending, pick city name, append only the first city to a list
l = [] # list to store cities with max temperature from each chunk
def set_max(c):
     x = c['temperature'].max()
     print(x)
     l.append(x)

        
colnames = ['city', 'temperature']
start_time = time.time()
for chunk in pd.read_csv('Final/measurements.txt', chunksize = 100000000, dtype={0:object, 1:float}, header = None, sep =';',names = colnames):
    set_max(chunk)
    
# Measure the end time and calculate the duration
end_time = time.time()
duration = end_time - start_time

# Print the results
print(f"Maximum temperature: {max(l)}")
print(f"Time taken: {duration} seconds")
```
